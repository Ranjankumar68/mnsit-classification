```python
# Import the main TensorFlow library
import tensorflow

# Import the Keras module from TensorFlow
from tensorflow import keras

# Import Sequential class for building a linear stack of layers
from tensorflow.keras import Sequential

# Import Dense (fully connected) and Flatten layers from Keras
from tensorflow.keras.layers import Dense, Flatten
```

```python
# Load the MNIST dataset of handwritten digits from Keras' datasets module
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
# Check the shape of the training data
X_train.shape
```

```python
# Check the shape of the test data
X_test.shape
```

```python
# y_train contains the labels (0-9) corresponding to each image in X_train
# Each label represents the digit shown in the corresponding training image
y_train
```

```python
# Import the matplotlib library for data visualization
import matplotlib.pyplot as plt

# Display the third image (index 2) in the X_train dataset using imshow
plt.imshow(X_train[2])
```

```python
# Access the first image in the training dataset
X_train[0]
```

```python
# Normalize the pixel values of the training images by dividing by 255
# This scales the pixel values to be between 0 and 1 (since the original range is 0-255)
X_train = X_train / 255.0

# Similarly, normalize the pixel values of the test images
X_test = X_test / 255.0
```

```python
# Access the first image in the training dataset after normalization
X_train[0]
```

```python
# Initialize a sequential model, which is a linear stack of layers
model = Sequential()

# Add a Flatten layer to convert each 28x28 input image into a 1D vector (784 elements)
# The input_shape=(28,28) specifies the shape of the input images
model.add(Flatten(input_shape=(28, 28)))

# Add a Dense (fully connected) layer with 128 units and ReLU activation function
# ReLU (Rectified Linear Unit) is commonly used for hidden layers to introduce non-linearity
model.add(Dense(128, activation='relu'))

# Add another Dense layer with 32 units and ReLU activation
# This is a smaller hidden layer to further process the features
model.add(Dense(32, activation='relu'))

# Add a Dense output layer with 10 units (for 10 classes, digits 0-9) and softmax activation
# Softmax converts the output into probability distributions over the 10 classes
model.add(Dense(10, activation='softmax'))
```

```python
# Print a summary of the model architecture
model.summary()
```

```python
# Compile the model with the specified loss function, optimizer, and evaluation metrics

# 'sparse_categorical_crossentropy' is used as the loss function for multi-class classification tasks
# It is suitable when the target labels are integers (e.g., 0 to 9 for MNIST)
model.compile(loss='sparse_categorical_crossentropy',
              optimizer='Adam',         # Adam optimizer is an adaptive optimizer that works well for most tasks
              metrics=['accuracy'])     # 'accuracy' metric will be used to evaluate the model's performance during training
```

```python
# Train the model on the training data for 25 epochs
# validation_split=0.2 means 20% of the training data will be used for validation during training

history = model.fit(X_train,               # Training data
                    y_train,               # Training labels
                    epochs=25,             # Number of epochs to train the model
                    validation_split=0.2)  # 20% of the training data will be used for validation
```

```python
# Use the trained model to predict class probabilities on the test data
y_prob = model.predict(X_test)
```

```python
# Convert predicted probabilities to class labels by selecting the index with the highest probability
y_pred = y_prob.argmax(axis=1)
```

```python
# Import accuracy_score function from sklearn.metrics to evaluate the model's performance
from sklearn.metrics import accuracy_score

# Calculate the accuracy by comparing the true labels (y_test) with the predicted labels (y_pred)
accuracy_score(y_test, y_pred)
```

```python
# Plot the training and validation loss over the epochs to visualize the model's performance

# Plot the loss for the training data
plt.plot(history.history['loss'])

# Plot the loss for the validation data
plt.plot(history.history['val_loss'])
```

```python
# Plot the training and validation accuracy over the epochs to visualize the model's performance

# Plot the accuracy for the training data
plt.plot(history.history['accuracy'])

# Plot the accuracy for the validation data
plt.plot(history.history['val_accuracy'])
```
